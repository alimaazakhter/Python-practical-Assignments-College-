```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Load csv

df = pd.read_csv("agri.csv")
print("Agriculture production dataset")
print(df)
```

```python
print("\n Dataset Information")
df.info()
```

```python
#Highest Crop Production

max_crop = df[df['production'] == df['production'].max()]
print("\n Crop with highest production")
print(max_crop)
```

```python
# 4. Calculate average rainfall

avg_rainfall = df['rainfall'].mean()
print('\n Average Rainfall', avg_rainfall)
```

```python
# Show crops grown in “North” region.
north_crop = df[df['region'].str.lower() == 'north']
print('Crops grown in the north region')
print(north_crop)
```

```python
# Filter crops with production > 5000 tons.

crop_prod = df[df['production'] > 5000]
print("\n Crops with production > 5000")
print(crop_prod)
```

```python
# Sort by production

sorted_df = df.sort_values(by = 'production', ascending = False)
print("Dataset sorted by production: ")
print(sorted_df)
```

```python
# Group data by region

grouped = df.groupby('region').mean(numeric_only=True)
print('\n Grouped by region (Average): ')
print(grouped)
```

```python
 # Display summary statistics.

print("\n Summary Statistics")
print(df.describe())
```

```python
# Display last 8 rows

print("\n Display last 8 rows: ")
df.tail(8)
```

```python
# Visualization: Production per crop (bar chart).

plt.figure()
plt.bar(df['crop'], df['production'])
plt.xlabel('Crop')
plt.ylabel('Production (tons)')
plt.title('Production per crop')
plt.show()
```

```python
# 12. Visualization: Rainfall distribution (histogram).

plt.figure()
plt.hist(df['rainfall'], bins = 5)
plt.xlabel('Rainfall')
plt.ylabel('Frequence')
plt.title(' Rainfall distribution ')
plt.show()
```
